```python
def count(n):
    while n > 0:
        print("before yield")
        yield n
        n -= 1
        print("after yield")
g = count(5)
print(g.send(None))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

```python
def calc(*number):
    sum = 0
    for n in number:
        sum += n * n
    return sum
calc(1, 2, 3)
```

```python
def mysum(l):
    if not l:
        return 0
    else:
        return l[0] + mysum(l[1:])
mysum([i for i in range(1, 101)])
    
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
factorial(3)
```
