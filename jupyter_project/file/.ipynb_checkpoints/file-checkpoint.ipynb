### 拼接路径 os.path.join(path1[,path2[,...]])
### 判断目录是否存在  os.path.exists(path)
### 创建目录 os.mkdir(name, mode=0o777)  不能创建多级目录
### 创建多级目录  os.makedirs(name, mode=0o777)
### 删除目录 os.rmdir(path) 删除没有文件的空目录,如果不是就用shutil模块的shutil.rmtree
### 遍历目录 os.walk(top[, topdown][,onerror][, followlinks]) 返回三个元素的元组（dirpath, dirnames, filenames）
### 删除文件 os.remove(path)
### 重命名文件和目录 os.rename(src, dst)
### 获取文件信息 os.stat(path) 返回一个对象

```python
l1 = [1, 4, 3, 6, 9, 0, 2]
for x in sorted(l1):
    print(x)
for y in sorted(l1, reverse=True):
    print(y)
print(l1)
```

```python
def count(n):
    while n > 0:
        print("before yield")
        yield n
        n -= 1
        print("after yield")
g = count(5)
print(g.send(None))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

```python
sum = 0
for i in range(0,101):
    sum = sum + i
print(sum)
    
```

```python
i = 0
sum = 0
while i <= 100:
    sum = sum + i
    i += 1
print(sum)
```

```python
def calc(*number):
    sum = 0
    for n in number:
        sum += n * n
    return sum
calc(1, 2, 3)
```

```python
def mysum(l):
    if not l:
        return 0
    else:
        return l[0] + mysum(l[1:])
mysum([i for i in range(1, 101)])
    
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
factorial(3)
```
